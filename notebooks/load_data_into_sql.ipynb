```python
import pandas as pd
import mysql.connector
```

```python
csv_file = r"C:\Users\DIPIKA VISHWAKARMA\Desktop\inventary_project\data\raw_data\inventory_forecasting.csv"
df = pd.read_csv(csv_file)
```

```python
df.columns = df.columns.str.strip()

# Rename columns to match SQL table
df.rename(columns={
    'Store ID': 'Store_ID',
    'Product ID': 'Product_ID',
    'Inventory Level': 'Inventory',
    'Units Sold': 'Units_Sold',
    'Units Ordered': 'Units_Ordered',
    'Demand Forecast': 'Demand_Forecast',
    'Weather Condition': 'Weather_Condition',
    'Holiday/Promotion': 'Holiday_Promotion',
    'Competitor Pricing': 'Competition'
}, inplace=True)
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="pink#123",  # <-- apna MySQL password
  database="inventory_project"
)

mycursor = mydb.cursor()
```

```python
for index, row in df.iterrows():
    sql = """
    INSERT INTO Raw_Inventory 
    (Date, Store_ID, Product_ID, Category, Region, Inventory, Units_Sold, Units_Ordered, Demand_Forecast, Price, Discount, Weather_Condition, Holiday_Promotion, Competition, Seasonality) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    
    # Instead of tuple(row), we manually extract columns in correct order
    values = (
        str(row['Date']),
        str(row['Store_ID']),
        str(row['Product_ID']),
        str(row['Category']),
        str(row['Region']),
        int(row['Inventory']),
        int(row['Units_Sold']),
        int(row['Units_Ordered']),
        float(row['Demand_Forecast']),
        float(row['Price']),
        float(row['Discount']),
        str(row['Weather_Condition']),
        int(row['Holiday_Promotion']),
        float(row['Competition']),
        str(row['Seasonality'])
    )

    mycursor.execute(sql, values)

    if index % 500 == 0:
        print(f"{index} rows inserted...")

mydb.commit()
print(f"âœ… Total {index+1} rows inserted.")

mycursor.close()
mydb.close()
```
